# Preprocessing Data

```python
import pandas as pd

web99_df = pd.read_csv("../../data/99.csv")
rumah123_df = pd.read_csv("../../data/rumah123.csv")

rumah123_df
```

### Menghapus format m2 (meter persegi)

```python
def remove_m2(df):
    df["building_area"] = (
        df["building_area"].replace(r"(\d+)\s*m²", r"\1", regex=True).astype(int)
    )
    df["land_area"] = (
        df["land_area"].replace(r"(\d+)\s*m²", r"\1", regex=True).astype(int)
    )


remove_m2(rumah123_df)
remove_m2(web99_df)

rumah123_df
```

### Format harga

```python
def convert_currency(value):
    value = value.replace("Rp", "").replace(",", ".").strip()  # Hapus Rp & ganti , ke .

    if "Miliar" in value:
        return float(value.replace("Miliar", "").strip()) * 1_000
    elif "Juta" in value:
        return float(value.replace("Juta", "").strip())
    else:
        return float(value) / 1_000_000  # Jika tidak ada satuan, bagi dengan juta


rumah123_df["price"] = rumah123_df["price"].apply(convert_currency)
web99_df["price"] = web99_df["price"].apply(convert_currency)

rumah123_df
```

### Memastikan data yang dipakai itu rumah

```python
rumah123_df = rumah123_df[
    ~rumah123_df["name"].str.contains("kos|apartemen|tanah", case=False, na=False)
]
web99_df = web99_df[
    ~web99_df["name"].str.contains("kos|apartemen|tanah", case=False, na=False)
]
```

### Konsistensi nama lokasi

```python
rumah123_df["location"] = rumah123_df["location"].str.title()
web99_df["location"] = web99_df["location"].str.title()
```

Export CSV

```python
rumah123_df.to_csv("../../data/rumah123_cleaned.csv", index=False)
web99_df.to_csv("../../data/99_cleaned.csv", index=False)
```
