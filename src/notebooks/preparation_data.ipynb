# Preprocessing Data

```python
import pandas as pd

surabaya_df = pd.read_csv("../../data/rumah123_surabaya.csv")
gresik_df = pd.read_csv("../../data/rumah123_gresik.csv")
sidoarjo_df = pd.read_csv("../../data/rumah123_sidoarjo.csv")

rumah123_df = pd.concat([surabaya_df, gresik_df, sidoarjo_df])
rumah123_df
```

```python
# rumah123_df.columns = [col.lower().strip().replace(" ", "_") for col in rumah123_df.columns]
```

### Menghapus format m2 (meter persegi)

```python
def remove_m2(df):
    df["building_area"] = (
        df["building_area"].replace(r"(\d+)\s*m²", r"\1", regex=True).astype(float)
    )
    df["land_area"] = (
        df["land_area"].replace(r"(\d+)\s*m²", r"\1", regex=True).astype(float)
    )


remove_m2(rumah123_df)

rumah123_df
```

### Format harga

```python
def convert_currency(value):
    value = value.replace("Rp", "").replace(",", ".").strip()  # Hapus Rp & ganti , ke .

    if "Miliar" in value:
        return float(value.replace("Miliar", "").strip()) * 1_000
    elif "Juta" in value:
        return float(value.replace("Juta", "").strip())
    else:
        return float(value) / 1_000_000  # Jika tidak ada satuan, bagi dengan juta


rumah123_df["price"] = rumah123_df["price"].apply(convert_currency)
rumah123_df
```

### Memastikan data yang dipakai itu rumah

```python
rumah123_df = rumah123_df[
    ~rumah123_df["name"].str.contains("kos|apartemen|tanah", case=False, na=False)
]
```

### Sesuaikan nilai pada kolom location

```python
rumah123_df['location'] = rumah123_df['location'].replace('Surabaya Kota, Surabaya', 'Surabaya')
rumah123_df['location'] = rumah123_df['location'].replace('Waru, Surabaya', 'Waru, Sidoarjo')
rumah123_df = rumah123_df[rumah123_df['location'] != 'Sidoarjo, Surabaya']

rumah123_df['location'] = rumah123_df['location'].replace('Gresik, Gresik', 'Gresik')
rumah123_df['location'] = rumah123_df['location'].replace('Gresik Kota, Gresik', 'Gresik')

rumah123_df['location'] = rumah123_df['location'].replace('Sidoarjo, Sidoarjo', 'Sidoarjo')
```

### Konsistensi nama lokasi

```python
rumah123_df["location"] = rumah123_df["location"].str.title()
```

### Export CSV

```python
rumah123_df = rumah123_df.astype({"price": int, "land_area": int, "building_area": int})
rumah123_df.info()
```

```python
rumah123_df.to_csv("../../data/rumah123_cleaned.csv", index=False)
```
